```python
import pandas as pd
import numpy as np
from scipy.sparse import csr_matrix
import sklearn
from sklearn.decomposition import TruncatedSVD
```

```python
book = pd.read_csv('BX-Books.csv', sep =';', error_bad_lines = False, encoding = 'latin-1')
book.head()
```

```python
book.columns = ['ISBN','Book-Title','Book-Author', 'Year-Of-Publication','Publisher','Image-URL-S','Image-URL-M','Image-URL-L']
```

```python
book_user = pd.read_csv('BX-Users.csv', sep =';', error_bad_lines = False, encoding = 'latin-1')
book_user.head()
```

```python
book_user.columns = ['User-ID', 'Location', 'Age']
```

```python
Rating = pd.read_csv('BX-Book-Ratings.csv', sep =';', error_bad_lines = False, encoding = 'latin-1')
Rating.head()
```

```python
Rating.columns = ['User-ID','ISBN','Book-Rating']
```

```python
#since book and Rating had common column ISBN we can use that
Book_Rating = pd.merge(Rating, book, on='ISBN')
Book_Rating.head()
```

```python
columns = ['Book-Author', 'Year-Of-Publication','Publisher','Image-URL-S','Image-URL-M','Image-URL-L']
Book_Rating = Book_Rating.drop(columns, axis=1)
```

```python
Book_Rating.head()
```

```python
Book_Rating = Book_Rating.dropna(axis=0, subset = ['Book-Title'])
```

```python
book_ratingCount = (Book_Rating.
     groupby(by = ['Book-Title'])['Book-Rating'].
     count().
     reset_index().
     rename(columns = {'Book-Rating': 'totalRatingCount'})
     [['Book-Title', 'totalRatingCount']]
    )
book_ratingCount.head()
```

```python
Rating_totalRatingCount = Book_Rating.merge(book_ratingCount, left_on = 'Book-Title', right_on = 'Book-Title', how = 'left')
Rating_totalRatingCount.head()
```

```python
pd.set_option('display.float_format', lambda x: '%.3f' % x)
print(book_ratingCount['totalRatingCount'].describe())
```

```python
print(book_ratingCount['totalRatingCount'].quantile(np.arange(.9, 1, .01)))
```

```python
Rating_threshold = 50
rating_popular_book = Rating_totalRatingCount.query('totalRatingCount >= @Rating_threshold')
rating_popular_book.head()
```

```python
combined = rating_popular_book.merge(book_user, left_on = 'User-ID', right_on = 'User-ID', how = 'left')
us_india_user_rating = combined[combined['Location'].str.contains("usa|india")]
us_india_user_rating=us_india_user_rating.drop('Age', axis=1)
us_india_user_rating.head()
```

```python
if not us_india_user_rating[us_india_user_rating.duplicated(['User-ID', 'Book-Title'])].empty:
    initial_rows = us_india_user_rating.shape[0]

    print('Initial dataframe shape {0}'.format(us_india_user_rating.shape))
    us_india_user_rating = us_india_user_rating.drop_duplicates(['User-ID', 'Book-Title'])
    current_rows = us_india_user_rating.shape[0]
    print('New dataframe shape {0}'.format(us_india_user_rating.shape))
    print('Removed {0} rows'.format(initial_rows - current_rows))
```

```python
us_india_user_rating_pivot = us_india_user_rating.pivot(index = 'Book-Title', columns = 'User-ID', values = 'Book-Rating').fillna(0)
us_india_user_rating_matrix = csr_matrix(us_india_user_rating_pivot.values)
```

```python
from sklearn.neighbors import NearestNeighbors

model_knn = NearestNeighbors(metric = 'cosine', algorithm = 'brute')
model_knn.fit(us_india_user_rating_matrix)
```

```python

query_index = np.random.choice(us_india_user_rating_pivot.shape[0])
distances, indices = model_knn.kneighbors(us_india_user_rating_pivot.iloc[query_index, :].values.reshape(1, -1), n_neighbors = 6)

for i in range(0, len(distances.flatten())):
    if i == 0:
        print('Recommendations for {0}:\n'.format(us_india_user_rating_pivot.index[query_index]))
    else:
        print('{0}: {1}, with distance of {2}:'.format(i, us_india_user_rating_pivot.index[indices.flatten()[i]], distances.flatten()[i]))
```

```python
us_india_user_rating_pivot2 = us_india_user_rating.pivot(index = 'User-ID', columns = 'Book-Title', values = 'Book-Rating').fillna(0)
```

```python
us_india_user_rating_pivot2.head()
```

```python
us_india_user_rating_pivot2.shape
```

```python
X = us_india_user_rating_pivot2.values.T
X.shape
```

```python
SVD = TruncatedSVD(n_components=12, random_state=17)
matrix = SVD.fit_transform(X)
matrix.shape
```

```python
import warnings
warnings.filterwarnings("ignore",category =RuntimeWarning)
corr = np.corrcoef(matrix)
corr.shape
```

```python
us_india_book_title = us_india_user_rating_pivot2.columns
us_india_book_list = list(us_india_book_title)
coffey_hands = us_india_book_list.index("The Green Mile: Coffey's Hands (Green Mile Series)")
print(coffey_hands)
```

```python
corr_coffey_hands  = corr[coffey_hands]
```
